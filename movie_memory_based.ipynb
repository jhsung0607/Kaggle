#0. movie-collaborate-memory-based
* https://grouplens.org/datasets/movielens/latest/

```python
import pandas as pd
import numpy as np
```

```python
PATH = 'data/ml-latest-small'
movies = pd.read_csv(f'{PATH}/movies.csv')
ratings = pd.read_csv(f'{PATH}/ratings.csv')

movies.shape, ratings.shape
```

#1. 데이터 전처리

```python
# 유저별 영화평점 행렬 만들기
# 'userId', 'movieId', 'rating'
ratings_matrix = ratings.pivot_table(values="rating", index="userId", columns="movieId")
ratings_matrix
```

```python
# ratings 에 영화 제목과 장르 추가하기
# merge로 ratings와 movie를 movieId로 결합
rating_movies = ratings.merge(movies)
rating_movies.head(2)
```

```python
# 컬럼명이 영화명으로 보이는지 확인
# ratings = ratings[['userId', 'movieId', 'rating']]
ratings_matrix = rating_movies.pivot_table(values="rating",
                                           index="userId", columns="title")
ratings_matrix
```

```python
# NaN 데이터 0으로 대체
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix
```

#2.사용자 기반 최근접 이웃 협업 필터링
* 행 인덱스가 uesr, 열이 item 이 되도록 만든 행렬로 유사도를 구함

```python
# sklearn.metrics.pairwise 의 sklearn.metrics.pairwise 가져오기
from sklearn.metrics.pairwise import cosine_similarity
# ratings_matrix 간 유사도 구하기
# df_user_sim
user_sim = cosine_similarity(ratings_matrix, ratings_matrix)
df_user_sim = pd.DataFrame(user_sim, index=ratings_matrix.index,
                           columns=ratings_matrix.index)
df_user_sim.head()
```

## 예시

```python
# 특정 userId num번과 비슷한 취향의 영화를 보는 userID는?
user_num = 1
sim_user = df_user_sim[user_num].nlargest(10).index[1]
sim_user
```

```python
# 특정 userId의 영화평점은?
rating_movies.loc[rating_movies["userId"] == user_num, ["title", "rating", "userId"]]
```

```python
# 유사도가 가장 높은 사용자가 본 영화의 평점 찾기
rating_movies.loc[rating_movies["userId"] == sim_user,
                  ["title", "rating", "userId"]].nlargest(10, "rating")
```

```python
# 가장 유사도가 높은 사용자가 본 영화 중에 해당 유저가 안 본 영화를 평점이 높은 순으로 추천한다면?
mid = set(rating_movies.loc[rating_movies["userId"].isin([sim_user]), "movieId"]) \
- set(rating_movies.loc[rating_movies["userId"].isin([user_num]), "movieId"])
# mid
```

```python
rating_movies[rating_movies["userId"].isin([sim_user]) & rating_movies["movieId"].isin(mid)].nlargest(10, "rating")
```

#3 .아이템 기반 최근접 이웃 협업 필터링

* 행 인덱스가 item, 열이 user 가 되도록 만든 행렬로 유사도를 구함

```python
ratings_matrix.T
```

```python
# 행렬의 유사도 측정을 하기 위해 행렬을 transpose함
item_sim = cosine_similarity(ratings_matrix.T, ratings_matrix.T)
df_item_sim = pd.DataFrame(item_sim, index=ratings_matrix.T.index,
                           columns=ratings_matrix.T.index)
```

```python
# 아이템간 유사도 측정
df_item_sim
```
